Motivating example:
    In reinforcement learning, you need to choose between the action generated by argmax the q value and 
    an action randomly selected. You want to choose the random action by a probablity of epsilon and the 
    learnt action with a probability of 1-epsilon.
    
Implementation:
    1. create a two elements vector
    
    
Trap:
    1. the input log_prob should be 2-D tensor
    1.2 the column_index generated by tf.multimonial should be squeezed
    2. the mask should be converted to int32 after it is converted to one_hot

```python
import tensorflow as tf
```

```python
batch_size = 3
num_choices = 2

epsilon = 0.5

epsilon_t  = tf.placeholder("float", None)

log_prob = [[tf.log(epsilon_t),tf.log(1-epsilon_t)]]

column_index = tf.multinomial(log_prob,batch_size)

column_index_squeeze =  tf.squeeze(column_index)

mask = tf.to_int32(tf.one_hot(column_index_squeeze,num_choices))



sess = tf.Session()
print sess.run(mask,feed_dict={epsilon_t:epsilon})
```

```python
# action_learnt holds the action for each sample in the bat
action_learnt = tf.constant([1]*batch_size)
action_random = tf.constant([2]*batch_size)

actions = tf.stack([action_learnt,action_random],axis=1)
print sess.run(actions)
```

```python
action_final =  tf.reduce_sum(mask*actions,1)

print sess.run(action_final,feed_dict={epsilon_t:epsilon})
```
