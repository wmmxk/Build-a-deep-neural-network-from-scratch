Scenario: when you run the condition random field tutorial, the shape of the\
    logits generated by the VGG net followed by upsampling does not match with 
    the true labels. It might raise more errors if you modify the padding of VGG.
    One work around is to resize the label images

To read the image, the name of the image is fed into the session by a tf.placeholder storing a tf.string

```python
import tensorflow as tf
```

```python
# Decode a JPG image and resize it to 299 by 299 using default method.
image_filename = './ims/cat.jpg'

image_filename_placeholder = tf.placeholder(tf.string)

image_tensor = tf.read_file(image_filename_placeholder)

image_tensor = tf.image.decode_jpeg(image_tensor, channels=3)

resized_image = tf.image.resize_images(image_tensor, [299, 299])

feed_dict_to_use = {image_filename_placeholder: image_filename}
```

```python
sess = tf.Session()
```

```python
image_v, resized_image_v = sess.run([image_tensor,resized_image],feed_dict_to_use)
```

```python
from PIL import Image
mask_back = Image.fromarray(image_v)
mask_back
```
